```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Main Monte Carlo Function
def monte_carlo_pi(n):
    x = np.random.rand(n)
    y = np.random.rand(n)
    inside_circle = np.sum(x**2 + y**2 <= 1.0)
    return 4 * (inside_circle / n)

# 2. Iteration Parameters
# Iterating from 10 to 1000 in steps of 20
sample_sizes = np.arange(10, 1001, 20)
actual_pi = np.pi
errors = []

# 3. Execution and Error Calculation
for n in sample_sizes:
    calculated_pi = monte_carlo_pi(n)
    # True error: |xtrue - xapproximate|
    error = abs(actual_pi - calculated_pi)
    errors.append(error)

# 4. Plotting Routine
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, errors, marker='o')
plt.xlabel('Number of Samples')
plt.ylabel('Absolute Error')
plt.title('Monte Carlo Pi Approximation Error')
plt.grid(True)

# 5. Show and Save Result
plt.show()
```
