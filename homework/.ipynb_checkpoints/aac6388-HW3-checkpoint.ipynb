```python
import numpy as np
import matplotlib.pyplot as plt
import math

# --- Setup ---
def sin_approx(x, num_terms):
    # Taylor series approximation for sin(x)
    approx = 0
    for n in range(num_terms):
        term = ((-1)**n) * (x**(2*n + 1)) / math.factorial(2*n + 1)
        approx += term
    return approx

# Domain: 0 to 10
x_vals = np.arange(0, 10.1, 0.1)
exact_vals = np.sin(x_vals)

# --- Calculation & Plotting ---
plt.figure(figsize=(10, 6))

for terms in [1, 2, 3, 4]:
    # Approximations for this specific number of terms
    y_approx = np.array([sin_approx(x, terms) for x in x_vals])
    
    # Absolute Truncation Error
    error = np.abs(exact_vals - y_approx)
    
    plt.plot(x_vals, error, label=f'{terms} Term(s)')

# --- Graph Details ---
plt.title('Truncation Error of Taylor Approximation for Sin(x)')
plt.xlabel('x (radians)')
plt.ylabel('Absolute Error')
plt.ylim(0, 5) # Limiting y-axis to see useful data
plt.legend()
plt.grid(True)
plt.show()
```
