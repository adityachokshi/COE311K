```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(n):
    x = np.random.rand(n)
    y = np.random.rand(n)
    inside = np.sum(x**2 + y**2 <= 1.0)
    return 4 * (inside / n)

sample_sizes = np.arange(10, 1001, 20)
actual_pi = np.pi
errors = []

for n in sample_sizes:
    calculated_pi = monte_carlo_pi(n)
    error = abs(calculated_pi - actual_pi)
    errors.append(error)

plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, errors)
plt.xlabel('Number of Samples')
plt.ylabel('Absolute Error')
plt.title('Monte Carlo Pi Approximation Error')
plt.grid(True)
plt.show()
```
