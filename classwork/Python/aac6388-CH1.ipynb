```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

rand_x = np.random.rand(500)
rand_y = np.random.rand(500)

data = np.random.randn(1000)
heatmap_data = np.random.rand(20, 20)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("sin(x)")


axs[0, 1].scatter(rand_x, rand_y)
axs[0, 1].set_title("Scatter Plot")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("y")


axs[1, 0].hist(data, bins=30)
axs[1, 0].set_title("Histogram")
axs[1, 0].set_xlabel("Value")
axs[1, 0].set_ylabel("Frequency")


im = axs[1, 1].imshow(heatmap_data, cmap="hot")
axs[1, 1].set_title("Heatmap")
fig.colorbar(im, ax=axs[1, 1])

plt.tight_layout()
plt.show()
```
