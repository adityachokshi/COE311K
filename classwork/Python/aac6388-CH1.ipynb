```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))

x2 = np.random.rand(500)
y2 = np.random.rand(500)
plt.subplot(2, 2, 2)
plt.scatter(x2, y2)

data = np.random.randn(1000)
plt.subplot(2, 2, 3)
plt.hist(data, 30)

z = np.random.rand(20, 20)
plt.subplot(2, 2, 4)
im = plt.imshow(z, cmap='hot')
plt.colorbar(im)

plt.show()
```
