```python
import numpy as np

# Initialize variables
pi_approx = 0
terms = 5000

# Header for the output table
print(f"{'n':<10} | {'Approx Pi':<15} | {'Relative Error':<20}")
print("-" * 50)

for n in range(terms):
    # Leibniz Formula: sum of 4 * (-1)^n / (2n + 1)
    numerator = (-1)**n
    denominator = 2*n + 1
    term = 4 * (numerator / denominator)
    
    # Add term to running total
    pi_approx += term
    
    # Check if we have hit a multiple of 100
    # We use n+1 because n starts at 0
    if (n + 1) % 100 == 0:
        true_val = np.pi
        rel_error = abs((pi_approx - true_val) / true_val)
        
        # Print with formatted strings (scientific notation for error)
        print(f"{n+1:<10} | {pi_approx:<15.6f} | {rel_error:<20.4e}")
```
