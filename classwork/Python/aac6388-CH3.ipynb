```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Estimation Function
def estimate_pi(n):
    x, y = np.random.rand(n), np.random.rand(n)
    inside = np.sum(x**2 + y**2 <= 1.0)
    return 4 * (inside / n)

# 2. Parameters and Setup
n_samples = 1000
n_runs = 100
true_pi = np.pi
abs_errors = []
rel_errors = []

# 3. Execution Loop and Error Calculation
for i in range(n_runs):
    calc_pi = estimate_pi(n_samples)
    
    # Absolute and Relative Error formulas
    abs_err = abs(true_pi - calc_pi)
    rel_err = abs((true_pi - calc_pi) / true_pi)
    
    abs_errors.append(abs_err)
    rel_errors.append(rel_err)

# 4. Data Preparation for Plotting
runs = np.arange(1, n_runs + 1)

# 5. Visualization
plt.figure(figsize=(10, 5))
plt.plot(runs, abs_errors, label='Absolute Error')
plt.plot(runs, rel_errors, label='Relative Error')
plt.xlabel('Run Number')
plt.ylabel('Error')
plt.title('Monte Carlo Pi Error Analysis (100 Runs)')
plt.legend()
plt.grid(True)
plt.show()
```
