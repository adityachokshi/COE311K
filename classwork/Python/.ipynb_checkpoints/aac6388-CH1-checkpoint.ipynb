```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x))
plt.show() 



```

```python
y = np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
plt.hist(x, 30)
plt.show()
```
